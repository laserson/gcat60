### Lab members

I manually copy `http://arep.med.harvard.edu/gclab6.htm` into an Excel document and save as tab-delimited data called `raw/members.txt`.  Then I have to pipe it to fix the newline issues:

    cat raw/members.txt | tr '\r' '\n' > raw/members.tsv

Finally, I manually edit a few entries that cause parsing problems (there are only a few).

```python
with open('raw/members.tsv', 'r') as ip:
    raw = ip.read()
raw = ''.join([c if ord(c) < 128 else '' for c in raw])
lines = [line.strip() for line in raw.split('\n')]
lines = filter(lambda x: x != '', lines)
data = [line.split('\t') for line in lines]

def to_record(fields):
    x = {}
    x['first_name'] = fields[0]
    x['last_name'] = fields[1]
    x['name'] = ' '.join([fields[0], fields[1]])
    x['start'] = parse_date(fields[3]).strftime('%Y-%m-%d')
    x['end'] = parse_date(fields[4]).strftime('%Y-%m-%d')
    x['tenure'] = (parse_date(x['end']) - parse_date(x['start'])).total_seconds()
    x['position_string'] = fields[6]
    return x

records = sorted(map(to_record, data), key=lambda x: (x['start'], x['end']))
with open('json/members.json', 'w') as op:
    for r in records:
        print >>op, json.dumps(r)
```

People over time

```python
# rows = records
rows = filter(lambda r: r['tenure'] > 31536000.0, records)

fig = plt.figure()
ax = fig.add_subplot(111)
for (i, row) in enumerate(rows):
    ax.plot([parse_date(row['start']), parse_date(row['end'])], [i+1, i+1])
ax.grid(True)
# ax.set_yscale('log')
ax.set_ylim(len(rows) + 10, -10)
```

Longest tenures

```python
def print_tenures(finished):
    for r in finished[:20]:
        duration_years = r['tenure'] / (60. * 60 * 24 * 30 * 12)
        print "%-20s %.2f" % (r['name'], duration_years)
```

```python
finished = filter(lambda r: r['end'] < '2014-07-01', records)
finished.sort(key=lambda r: -r['tenure'])
print_tenures(finished)
```

```python
finished_postdocs = filter(lambda x: 'P' in x['position_string'], finished)
print_tenures(finished_postdocs)
```

```python
finished_grads = filter(lambda x: 'G' in x['position_string'], finished)
print_tenures(finished_grads)
```

```python
finished_undergrads = filter(lambda x: 'U' in x['position_string'], finished)
print_tenures(finished_undergrads)
```

Shortest tenures

```python
print_tenures(finished_grads[::-1])
```

```python
one_month = datetime.timedelta(weeks=4)
start = parse_date(min([r['start'] for r in records]))
end = parse_date(max([r['end'] for r in records]))

dates = []
population = []
date = start
while date < end:
    def date_predicate(r):
        return True if (parse_date(r['start']) <= date and parse_date(r['end']) >= date) else False
    dates.append(date)
    population.append(len(filter(date_predicate, records)))
    date += one_month
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(dates, population, lw=2)
ax.set_xticks([datetime.datetime(x, 1, 1) for x in range(1985, 2015)])
ax.set_xticklabels([x for x in range(1985, 2015)], rotation=45)
# ax.set_yscale('log')
ax.grid()
```

```python
import numpy as np
def print_stats(x):
    print "size: %i" % len(x)
    print "min: %.2f" % min(x)
    print "max: %.2f" % max(x)
    print "mean: %.2f" % np.mean(x)
    print "stddev: %.2f" % np.std(x)
    print "median: %.2f" % np.median(x)
```

```python
dur_years = [f['tenure'] / (60. * 60 * 24 * 30 * 12) for f in finished_postdocs]
print_stats(dur_years)
fig = plt.figure()
ax = fig.add_subplot(111)
g = ax.hist(dur_years, bins=range(-2, 12))
```

```python
dur_years = [f['tenure'] / (60. * 60 * 24 * 30 * 12) for f in finished_grads]
print_stats(dur_years)
fig = plt.figure()
ax = fig.add_subplot(111)
g = ax.hist(dur_years, bins=range(-2, 12))
```
