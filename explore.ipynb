### Calendar time monopolization

Get some proxies for names

```python
first_names = {}
last_names = {}
for member in members:
    first_names[member['first_name']] = first_names.get(member['first_name'], 0) + 1
    last_names[member['last_name']] = last_names.get(member['last_name'], 0) + 1
names = list(set([x['name'] for x in members]))
proxies = {}
for member in members:
    proxies[member['name']] = [member['name']]
    if first_names[member['first_name']] == 1: proxies[member['name']].append(member['first_name'])
    if last_names[member['last_name']] == 1: proxies[member['name']].append(member['last_name'])
```

```python
time_accum = {}
for event in cal:
    if 'duration' not in event:
        continue
    for ne in event['names']:
        time_accum[ne.lower()] = time_accum.get(ne.lower(), 0.) + event['duration'] / 3600.
```

By full name only!

```python
time_by_member = {}
for member in members:
    time_by_member[member['name']] = time_accum.get(member['name'].lower(), 0)
```

```python
for p in sorted(filter(lambda p: p[1] is not None, time_by_member.iteritems()), key=lambda p: -p[1])[:21]:
    if p[1] > 0:
        print "%-20s%.2f" % p
```

Using proxies!

```python
time_by_member = {}
for (name, nicks) in proxies.iteritems():
    time_by_member[name] = time_accum.get(name.lower(), 0)
    for nick in nicks:
        time_by_member[name] += time_accum.get(nick.lower(), 0)
```

```python
for p in sorted(filter(lambda p: p[1] is not None, time_by_member.iteritems()), key=lambda p: -p[1])[:21]:
    if p[1] > 0:
        print "%-20s%.2f" % p
```

Now for organizations

Using NER

```python
time_by_org = {}
for org in orgs:
    time_by_org[org] = time_accum.get(org.lower(), 0)
```

```python
for p in sorted(filter(lambda p: p[1] is not None, time_by_org.iteritems()), key=lambda p: -p[1]):
    if p[1] > 0:
        print "%-20s%.2f" % p
```

Using manual string search

```python
time_by_org = {}
for org in orgs:
    time_by_org[org] = 0
    for event in cal:
        if 'duration' in event:
            if org.lower() in event['text'].lower():
                time_by_org[org] += event['duration'] / 3600.
```

```python
for p in sorted(filter(lambda p: p[1] is not None, time_by_org.iteritems()), key=lambda p: -p[1])[:20]:
    if p[1] > 0:
        print "%-20s%.2f" % p
```

### Correlations

```python
pubs_by_member = {}
for pub in pubs:
    for author in pub['authors']:
        if 'church' in author.lower():
            continue
        pubs_by_member[author.lower()] = pubs_by_member.get(author.lower(), 0) + 1
```

```python
lm_by_member = {}
for member in members:
    if 'church' in member['last_name'].lower():
        continue
    lm_by_member[member['last_name'].lower()] = 0
    for meeting in lm:
        if member['last_name'].lower() in meeting['text'].lower():
            lm_by_member[member['last_name'].lower()] += 1
```

```python
x = [lm_by_member[member['last_name'].lower()] for member in members]
y = [pubs_by_member.get(member['last_name'].lower(), 0) for member in members]
points = filter(lambda p: max(p) < 50, zip(x, y))
(x, y) = zip(*points)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
ax.set_xlabel('lab meetings')
ax.set_ylabel('publications')
```
